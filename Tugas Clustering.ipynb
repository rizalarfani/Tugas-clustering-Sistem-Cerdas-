```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans


dataset = pd.read_csv("Mall_Customers.csv")
dataset_x = dataset.iloc[:,3:5]
sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=dataset, s=100, color="red", alpha = 0.5)
dataset_array = np.array(dataset_x)
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(dataset_array)
#Menentukan dan mengkonfigurasi fungsi kmeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter =300, n_init = 10, random_state = 0)
#Menentukan kluster dari data
kmeans.fit_predict(x_scaled)
# # Menambahkan kolom "kluster" dalam data frame ritel
dataset["kluster"] = kmeans.labels_
fig, ax = plt.subplots()
sct = ax.scatter(x_scaled[:,1], x_scaled[:,0], s = 100,
c = dataset.kluster, marker = "o", alpha = 0.5)
centers = kmeans.cluster_centers_
ax.scatter(centers[:,1], centers[:,0], c='blue', s=200, alpha=0.5)
ax.scatter(centers[:,1], centers[:,0], s = 300, c = 'blue',  alpha=0.5, label = 'Centroids')
plt.title("Hasil Klustering K-Means")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
